# A Python Package for Parameter Estimation and Uncertainty Quantification


This jupyter notebook accompanies my talk at the [COMBINE conference 2020](http://co.mbine.org/events/COMBINE_2020). 

[PyPESTO](https://github.com/ICB-DCM/pyPESTO) is a python package for parameter inference. PyPESTO provides an interface to the model simulation tool [AMICI](https://github.com/AMICI-dev/AMICI) and hence is tailored, but not restricted, to Ordinary Differential Equation (ODE) models specified in the [SBML](https://github.com/ICB-DCM/pyPESTO) & [PEtab](https://github.com/PEtab-dev/PEtab) standard.

Mathematically, this talk focus on ODE models of the form
\begin{align*}
\dot{x} &= f(x, \theta), \;\;\; x_0=x_0(\theta) \\
d_i &= h(x(t_i, \theta), \theta) + \varepsilon_i
\end{align*}

With i.i.d. measurement noise $\varepsilon_i$. Parameter estimates $\hat{\theta}$ are commonly derived via 
maximizing a likelihood or posterior distribution.

```python
import pypesto
import pypesto.petab 
import pypesto.optimize as optimize
import pypesto.visualize as visualize

import petab
import amici
import os.path
import numpy as np
```

## PEtab

The easiest way to define and import a parameter fitting problem in pyPESTO is by importing a PEtab problem. PEtab is a data format for specifying parameter estimation problems in systems biology. 

<img src="img/petab_concept.png" width="100%" alt="pyPESTO logo"/>

A PEtab problem consist of an SBML file, defining the model topology and a set of `.tsv` files, defining experimental conditions, observables, measurements and parameters (and their optimization bounds, scale, priors...). All files, that make up a PEtab problem, can get structured in a problem `.yaml` file.

For more details on PEtab, the interested reader is refered to [PEtab's format definition](https://github.com/PEtab-dev/PEtab/blob/master/doc/documentation_data_format.rst), for examples, the reader is refered to the [PEtab benchmark collection](https://github.com/Benchmarking-Initiative/Benchmark-Models-PEtab). The Model by _[BÃ¶hm et al. JProteomRes 2014](https://pubs.acs.org/doi/abs/10.1021/pr5006923)_ is part of the benchmark collection and will be used as the running example throughout this notebook.

## Basic Model Import and Optimization 

The first step is to import a PEtab problem and create a `pypesto.problem` object:

```python
# directory of the PEtab problem
petab_yaml = './Boehm_JProteomeRes2014/Boehm_JProteomeRes2014.yaml' 

importer = pypesto.petab.PetabImporter.from_yaml(petab_yaml)
problem = importer.create_problem()
```

Next, we choose an `optimizer` and a number of optimization runs (`n_starts`) to perform the multi start optimization.

```python
%%time
%%capture

optimizer = optimize.ScipyOptimizer()

# do the optimization
result = optimize.minimize(problem=problem, 
                           optimizer=optimizer,
                           n_starts=10)
```

`result.optimize_result` now stores as list, that contains the results and meta data of the individual optimizer runs (ordered by function value).

```python
# E.g. best model fit was obtained by the following optimization run:
result.optimize_result.list[0]
```

```python
# Objective function values of the different optimizer runs:
result.optimize_result.get_for_key('fval')
```

## Optimizer Choice:

PyPESTO provides a unified interface to a variety of optimizers of different types:

* All [scipy optimizer](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html#scipy.optimize.minimize) (`optimize.ScipyOptimizer(method=<method_name>)`)
    * _function-value_ or _least-squares_-based optimizers
    * _gradient_  _hessian_-based optimizers
* [IpOpt](https://pypi.org/project/ipopt/) (`optimize.IpoptOptimizer()`)
    * Interior point method
* [Dlib](http://dlib.net) (`optimize.DlibOptimizer(options={'maxiter': <max. number of iterations>})`)
    * Globa optimizer
    * gradient free
* [Particle Swarm](https://github.com/tisimst/pyswarm) (`optimize.PyswarmOptimizer()`)
    * Particle swarm algorithm
    * gradient free
* [CMA-ES](https://pypi.org/project/cma-es/) (`optimize.CmaesOptimizer()`)
    * Covariance Matrix Adaptation Evolution Strategy
    * Evolutionary Algorithm

```python
optimizer_scipy_lbfgsb = optimize.ScipyOptimizer(method='L-BFGS-B')
optimizer_scipy_powell = optimize.ScipyOptimizer(method='Powell')

optimizer_dlib = optimize.DlibOptimizer(method = '', options={'maxiter': 10000})
optimizer_pyswarm = optimize.PyswarmOptimizer()
```

The following performs 100 multi-start runs with different optimizers in order to compare their performance. 

_Note_: `dlib` and `pyswarm` need to be installed for this section to run. Furthermore the computation time is in the order of minutes, so you might want to skip the excecution and jump to the next section.  

```python
%%time
%%capture
n_starts = 100

# Due to run time we here already use parallelization, 
# which will be introduced in more detail later
engine = pypesto.engine.MultiProcessEngine()
history_options = pypesto.HistoryOptions(trace_record=True)

# Scipy: L-BFGS-B
result_lbfgsb = optimize.minimize(problem=problem, 
                                  optimizer=optimizer_scipy_lbfgsb,
                                  engine=engine,
                                  history_options=history_options,
                                  n_starts=n_starts)

# Scipy: Powell
result_powell = optimize.minimize(problem=problem, 
                              optimizer=optimizer_scipy_powell,
                              engine=engine,
                              history_options=history_options,
                              n_starts=n_starts)

# Dlib
result_dlib = optimize.minimize(problem=problem, 
                                optimizer=optimizer_dlib,
                                engine=engine,
                                history_options=history_options,
                                n_starts=n_starts)


# PySwarm
result_pyswarm = optimize.minimize(problem=problem, 
                                   optimizer=optimizer_pyswarm,
                                   engine=engine,
                                   history_options=history_options,
                                   n_starts=n_starts) 
```

## Fitting for large scale models

When fitting large scale models (i.e. with >100 parameters and accordingly also more data), two important issues are efficient gradient computation and parallelization.

### Efficient gradient computation

As seen in the example above and as can be confirmed from own experience: If fast and reliable gradients can be provided, gradients based optimizers are favourable with respect to omptimizer convergence and run time.  

It has been shown, that adjoint sensitivity analysis propose a fast and reliable method to compute gradients for large scale models, since their run time is (asymptotically) independent of the number of parameters ([FrÃ¶hlich et al. PlosCB 2017](https://journals.plos.org/ploscompbiol/article/file?id=10.1371/journal.pcbi.1005331&type=printable)).

<img src="img/adjoint_sensitivities.png" width="40%" alt="pyPESTO logo"/> 

(Figure from FrÃ¶hlich et al. PlosCB 2017) Adjoint sensitivities are implemented in AMICI.

```python
# Set gradient computation method to adjoint
problem.objective.amici_model.setSteadyStateSensitivityMode(amici.SensitivityMethod.adjoint)
```

### Parallelization

Multi-start optimization can easily be parallelized by using `engines`.

```python
%%time
%%capture
# Parallelize
engine = pypesto.engine.MultiProcessEngine()

# Optimize
result = optimize.minimize(problem=problem, 
                           optimizer=optimizer_scipy_lbfgsb,
                           engine=engine,
                           n_starts=100)
```

# Objective Function Definition

PyPESTO is not restricted to ODE models, but allows to define objective functions. The following example demonstrates this functionality.

## Elastic Net regularization:

An [elastic net](https://en.wikipedia.org/wiki/Elastic_net_regularization) penalizes parameters by adding the penalty term
\begin{align}
\lambda_1 \cdot ||\theta ||_1 + \lambda_2 \cdot ||\theta ||_2^2
\end{align}
to the objective function. 

From a Bayesian perspective, L1 and L2 penalties can be interpreted as Laplace and Gaussian priors on $\theta$. Laplace and Gaussian priors are supported by PEtab and can be processed via `pypesto.petab.PetabImporter`. 

Here we want to take another approach and define a seperate `pypesto.Objective`, that contains the penalty terms. Therefore we define an objective function `elastic_net` and its gradient. 

(**Caution:** The PEtab problem specifies log10-transformed parameters. Therefore we have to re-transform parameters to the linear space, when evaluating the `elasitc_net` and its gradient.)

```python
import numpy.linalg as la

lambda_1 = 0
lambda_2 = 0

# define regularization
def elastic_net(x_log: np.ndarray):
    """elastic net for x specified in log-space"""
    # x_log = problem.get_reduced_vector(x_log)
    x_lin = 10**x_log
    return lambda_1 * la.norm(x_lin, ord=1) + lambda_2 * la.norm(x_lin)**2

# define gradient
def grad_elastic_net(x_log: np.ndarray):
    """gradient of elastic net for x specified in log-space"""
    # x_log = problem.get_reduced_vector(x_log)
    x_lin = 10**x_log
    grad_lin = lambda_1 * np.sign(x_lin) + 2 * x_lin
    return np.log(10) * la.multiply(grad_lin, x_log) # chain rule

# define Objective...
penalty_objective = pypesto.Objective(fun=elastic_net, grad=grad_elastic_net)
```

# Uncertainty quantification

PyPESTO focuses on two approaches to assess parameter uncertainties:

* Profile likelihoods
* Sampling

## Profile Likelihoods

[Profile likelihoods](https://academic.oup.com/bioinformatics/article/25/15/1923/213246) compute confidence intervals via a [likelihood ratio test](https://en.wikipedia.org/wiki/Likelihood-ratio_test). Profile likelihoods perform a maximum-projection of the likelihood function on the parameter of interest. The likelihood ratio test then gives a cut off criterium via the $\chi^2_1$ distribution.

In pyPESTO, the maximum projection is solved as maximization problem and can be obtained via

```python
%%time
%%capture

import pypesto.profile as profile

result = profile.parameter_profile(problem=problem, 
                                   result=result, 
                                   optimizer=optimizer_scipy_lbfgsb)
```

The maximum projections can now be inspected via:

```python
# adapt x_labels..
x_labels = [f'Log10({name})' for name in problem.x_names]

visualize.profiles(result, x_labels = x_labels, show_bounds=True)
```

## Sampling

In pyPESTO, sampling from the posterior distribution can be performed as 

```python
import pypesto.sample as sample

n_samples = 10000

sampler = sample.AdaptiveMetropolisSampler()

result = sample.sample(problem, 
                       n_samples=n_samples, 
                       sampler=sampler, 
                       result=result)
```

Sampling results are stored in `result.sample_result` and can be accessed e.g. via

```python
result.sample_result['trace_x']
```

### Samlping Diagnostics

Gewekes test accesses convergence of a sampling run and computes the burn in of a sampling result. The effective sample size indicates the strenght the correlation between different samples.

```python
sample.geweke_test(result=result)
result.sample_result['burn_in']
```

```python
sample.effective_sample_size(result=result)
result.sample_result['effective_sample_size']
```

### Visualization of Sampling Results

```python
# scatter plots
ax = visualize.sampling_scatter(result)

# marginals
ax = visualize.sampling_1d_marginals(result)
```
